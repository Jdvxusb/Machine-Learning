```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
yasListesi = list(range(10,100,10))
```

```python
kiloListesi=list(range(20,200,20))
```

```python
numpyYasListesi=np.array(yasListesi)
numpyKiloListesi=np.array(kiloListesi)
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi)
plt.xlabel("Yas")
plt.ylabel("Kilo")
plt.title("Kilonun yasa gore degisimi")
```
